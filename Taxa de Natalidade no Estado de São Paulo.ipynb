###### Analisando a base de dados do DataSUS, os 10 maiores indices da taxa de natalidade por mães menores de 19 anos.

As informações referentes à taxa de natalidade, selecionamos a sessão "informações de saúde", na área "estatísticas vitais" e o grupo de "nascidos vivos 1996 à 2016" restringindo mães menores de 19 anos nos anos de 2014, 2015 e 2016.
http://www2.datasus.gov.br/DATASUS/index.php?area=0205&id=6936 acesso em 15/11/2018.

Após a leitura dos dados, realizamos o download dos anos e concatenamos as informações e formatamos no padrão .CSV (Comma-separeted values - Valores separados por virgulas) e estruturamos da seguinte maneira


```python
# Importando as bibliotecas.
import pandas as pd
import seaborn as sb
low_memory=False
%matplotlib inline

# Utilizamos as variáveis "filename" e "TXN" para ler e armazenar os dados do arquivo ".csv".  
filename= 'Taxa_de_natalidade.csv'
TXN = pd.read_csv(filename, encoding='latin-1')

# Os itens foram selecionados para demonstrar os valores e destacar as 10 cidades do 
# estado de São Paulo com maior índice da taxa de natalidade, separados e organizados por colunas.                                
TXN.sort_values(by= ["Soma de Menor de 10 anos 2014", "Soma de 10 a 14 anos 2014", "Soma de 15 a 19 anos 2014", "Soma de Menor de 10 anos 2015", "Soma de 10 a 14 anos 2015", "Soma de 15 a 19 anos 2015", "Soma de Menor de 10 anos 2016", "Soma de 10 a 14 anos 2016", "Soma de 15 a 19 anos 2016"],ascending=False).head(10)
```

```python
# As mesmas informações à cima, serão demonstradas agora em gráficos de barras horizontais.
p = TXN.sort_values(by= ["Soma de Menor de 10 anos 2014", "Soma de 10 a 14 anos 2014", "Soma de 15 a 19 anos 2014", "Soma de Menor de 10 anos 2015", "Soma de 10 a 14 anos 2015", "Soma de 15 a 19 anos 2015", "Soma de Menor de 10 anos 2016", "Soma de 10 a 14 anos 2016", "Soma de 15 a 19 anos 2016"],ascending=False).head(10).plot(kind="barh", grid=(True), xticks=(100,1000,2000,3000,19000,21000,23000), x=("Município"),figsize=(20,10), fontsize=(16))
p.set_xlabel("Taxa de Natalidade", fontsize=(16));
p.set_title("Taxa de Natalidade por Mães menores de 19 anos, por municípios de São Paulo", fontsize=(20));
p.set_ylabel("Município", fontsize=(16));
```

Podemos observar que São Paulo é o município com maior índice de taxa de natalidade, seguido de Guarulhos e Campinas. São José dos Campos aparece no gráfico como nono município com maior taxa de natalidade.

Nossa análise acerca dos municípios do Estado de São Paulo demonstra maior índice da taxa de natalidade no município de São Paulo devido ser a maior densidade populacional do Estado e também sua capital, possuindo uma estimativa de 12.110.000 habitantes, sendo Guarulhos o segundo maior município em densidade populacional, contendo uma estimativa de 1.365.000 de habitantes.

```python
p = TXN.sort_values(by= ["Soma de 10 a 14 anos 2014", "Soma de 15 a 19 anos 2014", "Soma de Menor de 10 anos 2015", "Soma de 10 a 14 anos 2015", "Soma de 15 a 19 anos 2015", "Soma de Menor de 10 anos 2016", "Soma de 10 a 14 anos 2016", "Soma de 15 a 19 anos 2016"],ascending=False).head(1).plot(kind="bar", grid=(True), yticks=(500,1000,19000,20000,21000,22000,23000), x=("Município"),figsize=(15,15), fontsize=(12))
p.set_ylabel("Taxa de Natalidade", fontsize=(10));
p.set_title("Taxa de Natalidade por Mães menores de 19 anos, por municípios de São Paulo", fontsize=(16));
```
